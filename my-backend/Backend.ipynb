```python
# Import all the requirements
from flask import Flask, jsonify, request
import pandas as pd
import joblib

# Initialize the app and load data
app = Flask(__name__)
df = pd.read_csv('College_data.csv')
model = joblib.load('best_model.pkl')

#@app.route('/',methods=['GET'])
#def home():
#    return jsonify({"message": "Welcome to the College Recommendation API!"})
@app.route('/', methods=['GET'])
def home():
    """
    Root endpoint providing an overview of available API endpoints and their usage.
    """
    return jsonify({
        "message": "Welcome to the College Recommendation API!",
        "endpoints": {
            "/filter": {
                "method": "POST",
                "description": "Filters the college data based on student input criteria.",
                "parameters": {
                    "State": "State of the college",
                    "Stream": "Academic stream (e.g., Science, Arts)",
                    "UG_Fee": "Maximum undergraduate fee",
                    "PG_Fee": "Maximum postgraduate fee",
                    "Rating": "Minimum overall rating",
                    "Academic": "Minimum academic rating",
                    "Accommodation": "Minimum accommodation rating",
                    "Faculty": "Minimum faculty rating",
                    "Infrastructure": "Minimum infrastructure rating",
                    "Placement": "Minimum placement rating",
                    "Social_Life": "Minimum social life rating"
                }
            },
            "/predict": {
                "method": "POST",
                "description": "Predicts the composite score for a college based on input features.",
                "parameters": {
                    "UG_Fee": "Undergraduate fee",
                    "Placement": "Placement rating",
                    "Academic": "Academic rating",
                    "Accommodation": "Accommodation rating",
                    "Faculty": "Faculty rating",
                    "Infrastructure": "Infrastructure rating",
                    "Social_Life": "Social life rating"
                }
            }
        }
    })

@app.route('/filter', methods=['POST'])
def filter_colleges():
    """
    Filter the dataset based on student input (state, stream, fee, and ratings).
    """
    data = request.get_json()
    
    # Extract the filter parameters from the request
    state = data.get('State')
    stream = data.get('Stream')
    ug_fee_max = data.get('UG_Fee')
    pg_fee_max = data.get('PG_Fee')
    min_overall_rating = data.get('Rating')
    min_academic_rating = data.get('Academic')
    min_accommodation_rating = data.get('Accommodation')
    min_faculty_rating = data.get('Faculty')
    min_infrastructure_rating = data.get('Infrastructure')
    min_placement_rating = data.get('Placement')
    min_social_rating = data.get('Social_Life')

    # Apply filters to the dataset
    filtered_data = df[
        (df['State'] == state) &
        (df['Stream'] == stream) &
        (df['UG_Fee'] <= ug_fee_max) &
        (df['PG_Fee'] <= pg_fee_max) &
        (df['Rating'] >= min_overall_rating) &
        (df['Academic'] >= min_academic_rating) &
        (df['Accommodation'] >= min_accommodation_rating) &
        (df['Faculty'] >= min_faculty_rating) &
        (df['Infrastructure'] >= min_infrastructure_rating) &
        (df['Placement'] >= min_placement_rating) &
        (df['Social_Life'] >= min_social_rating)
    ]

    return jsonify({
        "message": "Filtered list of colleges based on provided criteria.",
        "filtered_colleges": filtered_data.to_dict(orient='records')
    })

@app.route('/predict', methods=['POST'])
def predict_colleges():
    """
    Predict and rank colleges based on composite score, using the ML model and student input.
    """
    data = request.get_json()

    # Filter data according to student's criteria, if any filters are specified
    filtered_data = df[
        (df['State'] == data.get('State', df['State'])) &
        (df['Stream'] == data.get('Stream', df['Stream'])) &
        (df['UG_Fee'] <= data.get('UG_Fee', df['UG_Fee'].max())) &
        (df['PG_Fee'] <= data.get('PG_Fee', df['PG_Fee'].max())) &
        (df['Rating'] >= data.get('Rating', df['Rating'].min())) &
        (df['Academic'] >= data.get('Academic', df['Academic'].min())) &
        (df['Accommodation'] >= data.get('Accommodation', df['Accommodation'].min())) &
        (df['Faculty'] >= data.get('Faculty', df['Faculty'].min())) &
        (df['Infrastructure'] >= data.get('Infrastructure', df['Infrastructure'].min())) &
        (df['Placement'] >= data.get('Placement', df['Placement'].min())) &
        (df['Social_Life'] >= data.get('Social_Life', df['Social_Life'].min()))
    ]

    # Prepare input features for prediction model
    input_features = filtered_data[['Rating', 'Placement', 'Academic', 'Accommodation', 'Faculty', 'Infrastructure', 'Social_Life']]

    # Predict scores and rank colleges
    filtered_data['Predicted Score'] = model.predict(input_features)
    ranked_colleges = filtered_data.sort_values(by='Predicted Score', ascending=False)

    return jsonify({
        "message": "Ranked colleges based on predicted composite score.",
        "ranked_colleges": ranked_colleges.to_dict(orient='records')
    })

if __name__ == '__main__':
    app.run(debug=True, use_reloader=False)

```

```python
from flask import Flask, jsonify, request
import pandas as pd
import joblib
from flask_pymongo import PyMongo

# Initialize the app and load data
app = Flask(__name__)

# MongoDB URI (Replace with your actual URI)
app.config["MONGO_URI"] = "mongodb://localhost:27017/college_data"  # Update with your URI
mongo = PyMongo(app)

# Load the dataset and model
df = pd.read_csv('College_data.csv')
model = joblib.load('best_model.pkl')

@app.route('/', methods=['GET'])
def home():
    """
    Root endpoint providing an overview of available API endpoints and their usage.
    """
    return jsonify({
        "message": "Welcome to the College Recommendation API!",
        "endpoints": {
            "/filter": {
                "method": "POST",
                "description": "Filters the college data based on student input criteria.",
                "parameters": {
                    "State": "State of the college",
                    "Stream": "Academic stream (e.g., Science, Arts)",
                    "UG_Fee": "Maximum undergraduate fee",
                    "PG_Fee": "Maximum postgraduate fee",
                    "Rating": "Minimum overall rating",
                    "Academic": "Minimum academic rating",
                    "Accommodation": "Minimum accommodation rating",
                    "Faculty": "Minimum faculty rating",
                    "Infrastructure": "Minimum infrastructure rating",
                    "Placement": "Minimum placement rating",
                    "Social_Life": "Minimum social life rating"
                }
            },
            "/predict": {
                "method": "POST",
                "description": "Predicts the composite score for a college based on input features.",
                "parameters": {
                    "UG_Fee": "Undergraduate fee",
                    "Placement": "Placement rating",
                    "Academic": "Academic rating",
                    "Accommodation": "Accommodation rating",
                    "Faculty": "Faculty rating",
                    "Infrastructure": "Infrastructure rating",
                    "Social_Life": "Social life rating"
                }
            }
        }
    })

@app.route('/store', methods=['POST'])
def store_data():
    """
    Store the student input data into MongoDB.
    """
    data = request.get_json()
    
    # Insert the student data into MongoDB
    mongo.db.student_data.insert_one(data)
    
    return jsonify({"message": "Data stored successfully!"}), 201

@app.route('/filter', methods=['POST'])
def filter_colleges():
    """
    Filter the dataset based on student input (state, stream, fee, and ratings).
    """
    data = request.get_json()

    # Fetch student data from MongoDB
    student_data = mongo.db.student_data.find_one({"_id": data.get('student_id')})

    if not student_data:
        return jsonify({"message": "Student data not found!"}), 404

    # Extract filter parameters from MongoDB data
    state = student_data['State']
    stream = student_data['Stream']
    ug_fee_max = student_data['UG_Fee']
    pg_fee_max = student_data['PG_Fee']
    min_overall_rating = student_data['Rating']
    min_academic_rating = student_data['Academic']
    min_accommodation_rating = student_data['Accommodation']
    min_faculty_rating = student_data['Faculty']
    min_infrastructure_rating = student_data['Infrastructure']
    min_placement_rating = student_data['Placement']
    min_social_rating = student_data['Social_Life']

    # Apply filters to the dataset
    filtered_data = df[
        (df['State'] == state) &
        (df['Stream'] == stream) &
        (df['UG_Fee'] <= ug_fee_max) &
        (df['PG_Fee'] <= pg_fee_max) &
        (df['Rating'] >= min_overall_rating) &
        (df['Academic'] >= min_academic_rating) &
        (df['Accommodation'] >= min_accommodation_rating) &
        (df['Faculty'] >= min_faculty_rating) &
        (df['Infrastructure'] >= min_infrastructure_rating) &
        (df['Placement'] >= min_placement_rating) &
        (df['Social_Life'] >= min_social_rating)
    ]

    return jsonify({
        "message": "Filtered list of colleges based on provided criteria.",
        "filtered_colleges": filtered_data.to_dict(orient='records')
    })

@app.route('/predict', methods=['POST'])
def predict_colleges():
    """
    Predict and rank colleges based on composite score, using the ML model and student input.
    """
    data = request.get_json()

    # Fetch student data from MongoDB
    student_data = mongo.db.student_data.find_one({"_id": data.get('student_id')})

    if not student_data:
        return jsonify({"message": "Student data not found!"}), 404

    # Extract features from MongoDB data
    state = student_data['State']
    stream = student_data['Stream']
    ug_fee_max = student_data['UG_Fee']
    pg_fee_max = student_data['PG_Fee']
    min_overall_rating = student_data['Rating']
    min_academic_rating = student_data['Academic']
    min_accommodation_rating = student_data['Accommodation']
    min_faculty_rating = student_data['Faculty']
    min_infrastructure_rating = student_data['Infrastructure']
    min_placement_rating = student_data['Placement']
    min_social_rating = student_data['Social_Life']

    # Filter the dataset based on student data
    filtered_data = df[
        (df['State'] == state) &
        (df['Stream'] == stream) &
        (df['UG_Fee'] <= ug_fee_max) &
        (df['PG_Fee'] <= pg_fee_max) &
        (df['Rating'] >= min_overall_rating) &
        (df['Academic'] >= min_academic_rating) &
        (df['Accommodation'] >= min_accommodation_rating) &
        (df['Faculty'] >= min_faculty_rating) &
        (df['Infrastructure'] >= min_infrastructure_rating) &
        (df['Placement'] >= min_placement_rating) &
        (df['Social_Life'] >= min_social_rating)
    ]

    # Prepare input features for prediction model
    input_features = filtered_data[['Rating', 'Placement', 'Academic', 'Accommodation', 'Faculty', 'Infrastructure', 'Social_Life']]

    # Predict scores and rank colleges
    filtered_data['Predicted Score'] = model.predict(input_features)
    ranked_colleges = filtered_data.sort_values(by='Predicted Score', ascending=False)

    return jsonify({
        "message": "Ranked colleges based on predicted composite score.",
        "ranked_colleges": ranked_colleges.to_dict(orient='records')
    })

if __name__ == '__main__':
    app.run(debug=True, use_reloader=False)
```
